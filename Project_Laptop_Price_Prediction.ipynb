```python
#default function is block of statements that can be called when we want it 
def fetch_processor(text):
    if text == 'Intel Core i7' or text == 'Intel Core i5' or text == 'Intel Core i3':
        return text
    else:
        if text.split()[0] == 'Intel':
            return 'Other Intel Processor'
        else:
            return 'AMD Processor'
a['Cpu brand'] = a['Cpu Name'].apply(fetch_processor)
a.sample(10)
```

```python
a['Memory'] = a['Memory'].astype(str).replace('\.0', '', regex=True) #replace  any ".0" values using regular expression with space
a["Memory"] = a["Memory"].str.replace('GB', '') #string replace 
a["Memory"] = a["Memory"].str.replace('TB', '000') #   ,, with 000
new = a["Memory"].str.split("+", n = 1, expand = True) 
#memory is seperated by "+" char and result is stored in new variable (n=1 [n will only occure once ,n is para])
#expand (it is a para and it set to "True" )


a["first"]= new[0] #create "first" column and store the new[0]
a["first"]=a["first"].str.strip() #removes any leading or trailing whitespace 
a["second"]= new[1] #create "second" column and stored the new[1]


a["Layer1HDD"] = a["first"].apply(lambda x: 1 if "HDD" in x else 0) #if HDD is present returns 1 otherwise 0, stored in l1HDD
a["Layer1SSD"] = a["first"].apply(lambda x: 1 if "SSD" in x else 0) #if SSD is present returns 1 otherwise 0, stored in l1SSD
a["Layer1Hybrid"] = a["first"].apply(lambda x: 1 if "Hybrid" in x else 0) #if Hybrid is present returns 1 otherwise 0,stoted in l1Hyrid 
a["Layer1Flash_Storage"] = a["first"].apply(lambda x: 1 if "Flash Storage" in x else 0) #if HDD is present returns 1 otherwise 0,stored in l2fla_Sto


a['first'] = a['first'].str.replace(r'\D', '') #replace '/d'(non digital charaters in regex)with space
a["second"].fillna("0", inplace = True) #missing value filled with 0 


a["Layer2HDD"] = a["second"].apply(lambda x: 1 if "HDD" in x else 0) #if HDD is present returns 1 otherwise 0, stored in l2HDD
a["Layer2SSD"] = a["second"].apply(lambda x: 1 if "SSD" in x else 0) #if SSD is present returns 1 otherwise 0, stored in l2SSD
a["Layer2Hybrid"] = a["second"].apply(lambda x: 1 if "Hybrid" in x else 0) #if Hybrid is present returns 1 otherwise 0,stoted in l2Hyrid 
a["Layer2Flash_Storage"] = a["second"].apply(lambda x: 1 if "Flash Storage" in x else 0) #if HDD is present returns 1 otherwise 0,stored in l2fla_Sto


a['second'] = a['second'].str.replace(r'\D', '') #replace '/d'(non digital charaters in regex)with space

a["first"] = a["first"].astype(int) #changing data type as 'int' 
a["second"] = a["second"].astype(int)

a["HDD"]=(a["first"]*a["Layer1HDD"]+a["second"]*a["Layer2HDD"]) #Each line is creating a new column by multiplying the first column and the corresponding layer column by each other
a["SSD"]=(a["first"]*a["Layer1SSD"]+a["second"]*a["Layer2SSD"]) #then adding the result to the product of the second column and the corresponding layer 2 column.
a["Hybrid"]=(a["first"]*a["Layer1Hybrid"]+a["second"]*a["Layer2Hybrid"])
a["Flash_Storage"]=(a["first"]*a["Layer1Flash_Storage"]+a["second"]*a["Layer2Flash_Storage"])

a.drop(columns=['first', 'second', 'Layer1HDD', 'Layer1SSD', 'Layer1Hybrid', #droping the unwated columns 
       'Layer1Flash_Storage', 'Layer2HDD', 'Layer2SSD', 'Layer2Hybrid',
       'Layer2Flash_Storage'],inplace=True)
```

**GET_DUMMIES ENCORDING**

```python
#get_dummies -> encoding use to convert string into numerical values
dummy=pd.get_dummies(a[['Company','TypeName','Cpu brand','Gpu brand','os']],drop_first=True)
dummy
```

**CONCATNATE**

```python
#cancatnate is to combine the datas 
b=pd.concat([a,dummy],axis=1)
b
```

```python
df=b.drop(['Company','TypeName','Cpu brand','Gpu brand','os','Weight'],axis=1)
df
```


**SPLTTING X AND Y**

```python
#splitting data into x and y
x = df.drop(columns=['Price']).values
y = (df['Price']).values
```

```python
x
```

```python
y
```

**TRAINING THE DATA**

```python
#set training data and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=1)
x_train
```

**LINEAR REGRESSION**

```python
#Model Creation
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
linear=r2_score(y_test,y_pred)
print('r2 score:',linear)
```

**RANDOM FOREST REGRESSOR**

```python
#Model Creation
from sklearn.ensemble import RandomForestRegressor
model2 = RandomForestRegressor()
model2.fit(x_train, y_train)
y_pred1=model2.predict(x_test)
```

```python
from sklearn.metrics import r2_score
RandomForest=r2_score(y_test,y_pred1)
print('r2 score:',RandomForest)
```

**DECISION TREE REGRESSOR**

```python
#Model Creation
from sklearn.tree import DecisionTreeRegressor
model3 = DecisionTreeRegressor()
model3.fit(x_train, y_train)
y_pred2=model3.predict(x_test)
```

```python
from sklearn.metrics import  r2_score
decisiontree=r2_score(y_test,y_pred2)
print('r2 score:',decisiontree)
```

**COMPARING THE r2_score OF ALL MODELS**

```python
models=['LINEAR REGRESSION','DECISON TREE','RANDOM FOREST']
r2=[linear,decisiontree,RandomForest]
scores= pd.DataFrame({'Model':models, 'R2_score':r2})
scores
```

```python
scores['R2_score']=scores['R2_score']*100
print(scores)
```

```python
#sorting models based on their accuracy score
scores=scores.sort_values(by='R2_score',ascending=True)
```



**COMPARING THE SCORES USING DIFFERENT GRAPHS**
