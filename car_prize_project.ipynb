#**Encoding Technique-Get Dummies**

```python
#get dummy encoding
dummy=pd.get_dummies(df[['Location','Fuel_Type','Transmission','Owner_Type']],drop_first=True)
dummy
```

```python
#drop unnecessary columns
dfe=df.drop(['Name','Location','Fuel_Type','Transmission','Owner_Type','New_Price','Unnamed: 0'],axis=1)
dfe
```

```python
#compaign encoding_done df with old df
df1=pd.concat([dfe,dummy],axis=1)
df1
```

##**Handling Missing Values**

```python
#clearing and filling missing values
df1.isna().sum()
```

###**String Replace**

```python
#string replace
#MILEAGE==km/kg and kmpl
df1['Mileage']=df1['Mileage'].str.replace('km/kg','')
df1['Mileage']=df1['Mileage'].str.replace('kmpl','')
df1['Engine']=df1['Engine'].str.replace('CC','')
df1['Power']=df1['Power'].str.replace('bhp','')
#null replace to zero
df1['Mileage']=df1['Mileage'].str.replace('null','0')
df1['Power']=df1['Power'].str.replace('null','0')
df1['Engine']=df1['Engine'].str.replace('null','0')
df1
```

```python
#print data type of each columns
df1.dtypes
```

```python
#convert dtype- object to float
df1['Engine']=df1['Engine'].astype(float)
df1['Power']=df1['Power'].astype(float)
df1['Mileage']=df1['Mileage'].astype(float)
```

```python
#check
df1.dtypes
```

```python
#missing values checking
df1.isna().sum()
```

```python
#(mileage engine power) of vehicles cant be zero
#so replace zero to a form
#IT is a  mathmetical opertn
#so import numpy 
import numpy as np
df1.loc[df1.Engine==0,'Engine']=np.NAN
df1.loc[df1.Power==0,'Power']=np.NAN
df1.loc[df1.Mileage==0,'Mileage']=np.NAN
```

```python
#check
df1.isna().sum()
```

```python
#fill missing values
df1['Engine']=df1['Engine'].fillna(df1['Engine'].mean())
df1['Power']=df1['Power'].fillna(df1['Power'].mean())
df1['Mileage']=df1['Mileage'].fillna(df1['Mileage'].mean())
df1['Seats']=df1['Seats'].fillna(df1['Seats'].mode()[0])
```

```python
df1.isna().sum()
```

#**Separate Input & Output datas**

```python
df2=df1.drop(['Price'],axis=1)
df2
```

```python
x=df2
y=df1['Price']
y
```

#**Train Test Split**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

#**Model Creation**

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
print(model.coef_)    #slope
print(model.intercept_) #y intercept
```

```python
af=pd.DataFrame({'AV':y_test,'PV':y_pred})
af
```

```python
plt.plot(y_test,color='y')
plt.plot(y_pred,color='r')
```

#**Performance Measurment**

```python
#performance measurment
from sklearn.metrics import mean_absolute_error
mean=mean_absolute_error(y_test,y_pred)
mean
```

```python
from sklearn.metrics import mean_absolute_percentage_error
mpe=mean_absolute_percentage_error(y_test,y_pred)
mpe
```

```python
from sklearn.metrics import mean_squared_error
error=mean_squared_error(y_test,y_pred)
error
```

```python
root=np.sqrt(error)
root
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
score
```
